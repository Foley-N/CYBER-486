1. Write a python program to create a plot to get general statistics of iris data

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('iris.csv')
df = pd.DataFrame(data)
df.drop('Id',axis=1).plot()
plt.show()
```

2. Write a Python program to create a Bar plot to get the frequency of the three
species of the Iris data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data =pd.read_csv("iris.csv",sep=",")
df = pd.DataFrame(data)
df['variety'].value_counts().plot.bar()

```

3. Write a Python program to create a Pie plot to get the frequency of the three
species of the Iris data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data =pd.read_csv("iris.csv",sep=",")
df = pd.DataFrame(data)
varietycountdf = df.groupby(['variety'])['variety'].count()
varietycountdf.plot.pie(legend=True, autopct='%1.0f')
```

4. Write a Python program to create a graph to find the relationship between the
sepal length and width

```python
import pandas as pd
import matplotlib.pyplot as plt

data =pd.read_csv("iris.csv",sep=",")
df = pd.DataFrame(data)
df['sepal.length'].plot(legend=True)
df['sepal.width'].plot(legend=True)
plt.show()
```

6. Write a Python program to find the correlation between variables of iris data.
Also create a heatmap using Seaborn to present their relations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv",sep=",")
df = pd.DataFrame(data)
df.drop('Id', axis=1)
df.drop('variety', axis=1)
corr = df.corr()
ax1 = sns.heatmap(corr, cbar=1, linewidths=2,vmax=3, vmin=0, square=True)
plt.show()



```
